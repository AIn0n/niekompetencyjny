```python
def newChild(parent1 :Thief, parent2 :Thief) -> list:
    point = random.choice(range(len(parent1.chromosome)))
    child1,child2 = Thief(), Thief()
    child1.setChromosome(parent1.chromosome[0:point] + parent2.chromosome[point:])
    child2.setChromosome(parent2.chromosome[0:point] + parent1.chromosome[point:])
    return [child1, child2]
```

```python
def buildNewGeneration(old_gen, elitarism, mut_prob):
    #elitarism
    old_gen.sort(key = lambda x : x.fitness, reverse = True)
    new_gen = [*old_gen[:int(len(old_gen)*elitarism)]]
    #looking for parents
    while len(new_gen) < len(old_gen):
        parents = random.choices(old_gen, weights = [x.fitness for x in old_gen], k = 2)
        children = newChild(*parents)
        for child in children:
            child.mutate(mut_prob)
        new_gen.extend(children)
    return new_gen
population = [Thief() for x in range(20)]

for x in range(10):
    population = buildNewGeneration(population, 0.2, 0.05)
    print("new gen")
    for thief in sorted(population, key = lambda x : x.fitness):
        print(thief)
```
